```python
def inputzolotoe():
   
    param = input('Введите название переменной ')
    if len(param) > 1:
        print('Введено более одной переменной')
    elif len(param) < 1:
        print('Введено менее одной прееменной')
    f = input('Функция в аналитическом виде :  ')
    a = input('Введите "a": ')
    b = input('Введите "b": ')

 
    Final = {'p1': param,
             'func': f,
             'a': a,
             'b': b}
    return Final
```

```python
from sympy import *
def zolotoe(data):
    import time 
    import numpy as np
    import sympy
    from matplotlib import pyplot as plt   
    from matplotlib import style   
    
    from sympy.parsing.sympy_parser import parse_expr

    f = data['func']
    f = parse_expr(f)
   
    a = float(data['a'])
    b = float(data['b'])
    #a,b = symbols('a b')
    
    start_time = time.time()
   
    e=0.05
    alpha=1.618
    x = symbols('x')
    n = round(np.log((-a+b)/e)/np.log(alpha))
    x1=b-(b-a)/alpha
    x2=a+(b-a)/alpha
    y1=f.subs(x,x1)
    y2=f.subs(x,x2)
    k=[]
    plt.plot(k,'b', label='line one', linewidth=5)   
    plt.title('значения функции ')   
    plt.rcParams['figure.figsize'] = (5,5)
    plt.xlabel('X axis')   
    plt.ylabel('y axis') 
    for i in range(1,n+1):
        
        
        if y1<=y2:
            b=x2
            x2=x1
            x1=b-(b-a)/alpha
            y2=y1
            y1=f.subs(x,x1)
            if abs(b-a)>=e:
                i=i+1
            elif abs(b-a)<e:
                print(f'x={a}')
                print(f'y={y1}')
                print(f'i(количество итераций цикла)={i}')
            k.append(y1)
            
            plt.plot(k,'b', label='line one', linewidth=5)  
            
        
               
        elif y1>y2:
            a=x1
            x1=x2
            x2=a+(b-a)/alpha
            y1=y2
            y2=f.subs(x,x2)
            if abs(b-a)>=e:
                i=i+1
            elif abs(b-a)<e:
                print(f'x-точка минимума={a}')
                print(f'y-значение функции в точки минимума={y1}')
                print(f'i(количество итераций цикла)={i}')
               
            k.append(y1)
            
            plt.plot(k,'b', label='line one', linewidth=5) 
            
           
        
      

   
        
            
    plt.scatter(n-1, y1, c='red',linewidth=10)
    print("время выполнения( в секундах ) = %s  " % (time.time() - start_time))
#     return {'x1' : x1, 
#             'x2' : x2,
#             'y1':y1,
#             'y2':y2,
#             'alpha':alpha,
#             'e':e }
```
